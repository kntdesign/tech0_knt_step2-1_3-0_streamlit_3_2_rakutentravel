```python
#必要なライブラリをインポートする
import requests
import json
import pandas as pd

#楽天トラベル地区コードAPIを取得する
REQUEST_URL_GetAreaClass = 'https://app.rakuten.co.jp/services/api/Travel/GetAreaClass/20131024?'
APP_ID = '1043594254102804262'

params = {
    'format':'json',
    'formatVersion': '2',
    "applicationId":APP_ID
}

res_GetAreaClass = requests.get(REQUEST_URL_GetAreaClass, params)
data = res_GetAreaClass.json()
```

```python
#下記に記載されている記事の内容を一部修正し、ファイルを作成
#https://python-yutaka.com/rakuten_api_get_area_class_list/

import json
import pandas as pd

with open('rakuten_api_chikucode_file.json') as f:
    res = json.load(f)

prefectures = res['areaClasses']['largeClasses'][0]['largeClass'][1]['middleClasses']

list_a_1 = []#middleClassCodeを入れるリスト
list_a_2 = []#middleClassNameを入れるリスト
list_b_1 = []#smallClassCodeを入れるリスト
list_b_2 = []#smallClassNameを入れるリスト
list_c_1 = []#detailClassCodeを入れるリスト
list_c_2 = []#detailClassNameを入れるリスト

for i in range(len(prefectures)):
    for j in range(len(prefectures[i]['middleClass'][1]['smallClasses'])):
        if len(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass']) == 2:
            for k in range(len(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass'][1]['detailClasses'])):
                list_a_1.append(prefectures[i]['middleClass'][0]['middleClassCode'] )
                list_a_2.append(prefectures[i]['middleClass'][0]['middleClassName'])
                list_b_1.append(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass'][0]['smallClassCode'] )
                list_b_2.append(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass'][0]['smallClassName'])
                list_c_1.append(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass'][1]['detailClasses'][k]['detailClass']['detailClassCode'])
                list_c_2.append(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass'][1]['detailClasses'][k]['detailClass']['detailClassName'])
        
        else:
            list_a_1.append(prefectures[i]['middleClass'][0]['middleClassCode'] )
            list_a_2.append(prefectures[i]['middleClass'][0]['middleClassName'])
            list_b_1.append(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass'][0]['smallClassCode'])
            list_b_2.append(prefectures[i]['middleClass'][1]['smallClasses'][j]['smallClass'][0]['smallClassName'])
            list_c_1.append(None)
            list_c_2.append(None)


df_a_1 = pd.DataFrame(list_a_1, columns = ['middleClassCode'])
df_a_2 = pd.DataFrame(list_a_2, columns = ['middleClassName'])
df_b_1 = pd.DataFrame(list_b_1, columns = ['smallClassCode'])
df_b_2 = pd.DataFrame(list_b_2, columns = ['smallClassName'])
df_c_1 = pd.DataFrame(list_c_1, columns = ['detailClassCode'])
df_c_2 = pd.DataFrame(list_c_2, columns = ['detailClassName'])



df_all = pd.concat([df_a_1, df_a_2, df_b_1, df_b_2,df_c_1,df_c_2], axis=1, ignore_index=False)

df_all.to_csv('rakuten_api_chikucode_ichiran.csv', index=False)
```
